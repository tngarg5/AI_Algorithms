
## DFS algorithm 

The DSF algorithm follows as:

We will start by putting any one of the graph's vertex on top of the stack.
After that take the top item of the stack and add it to the visited list of the vertex.
Next, create a list of that adjacent node of the vertex. Add the ones which aren't in the visited list of vertexes to the top of the stack.
Lastly, keep repeating steps 2 and 3 until the stack is empty.

## Create tree in python 


```python
#using dictionary to create a tree

graph = {
  '5' : ['3','7'],
  '3' : ['2', '4'],
  '7' : ['8'],
  '2' : [],
  '4' : ['8'],
  '8' : []
}

g2={
    'a' : ['c'],
    'c' : ['b','e','d'],
    'b' : ['e','c'],
    'd': ['c'],
    'e':['c','b'],
    'f':[]
}

#using set() function to set a visited node: 
#The set() function creates a set object. The items in a set list are unordered,
# so it will appear in random order. 
visited = set()

def dfs(visited,graph,node):
  #if node is not visited we print it and add it to the visited 
  if node not in visited:
    print(node)
    visited.add(node)
    #traversing each values of node from graph
    for adjacent in graph[node]:
      dfs(visited,graph,adjacent)


  #drivers code
print('\n 1st Example') 
dfs(visited,graph,'5')
print('\n 2nd Example')
dfs(visited,g2,'a')
```
